```python
import pandas as pd
import numpy as np
```

```python
mydf=pd.read_csv(r'imdb_labelled.txt',delimiter='\t', header=None)
```

```python
mydf.columns=['Review','Sentiment']
mydf.head()
# 0-->Negative
# 1-->Positive
```

```python
mydf.shape
```

```python
pd.options.display.max_colwidth=1000
```

```python
mydf = mydf.applymap(lambda s:s.lower().strip() if type(s) == str else s)
mydf.head()
```

```python
mydf.info()
```

```python
X=mydf.values[:,0]
Y=mydf.values[:,1]
Y=Y.astype(int)
```

```python
print(X)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer()
cv.fit(X)
X = cv.transform(X)
```

```python
print(X)
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=10)
```

```python
from sklearn.naive_bayes import BernoulliNB
from sklearn.naive_bayes import MultinomialNB

model=MultinomialNB(alpha=1.0)
#model=BernoulliNB(alpha=1.0, binarize=0)
#Alpha-->alue for Laplacian correction method to avoid zero frequency problems
model.fit(X_train,Y_train)

Y_pred= model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
cfm=confusion_matrix(Y_test,Y_pred)
print("Confusion matrix ",cfm)
print()
class_=classification_report(Y_test,Y_pred)
print("Classfication report ",class_)
print()
acc=accuracy_score(Y_test,Y_pred)
print("Accuracy score ",acc)
```
